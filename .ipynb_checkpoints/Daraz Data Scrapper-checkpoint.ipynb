Apply Vader Sentiment to find the sentiment of the review, and append it in a column named "Sentiment".

You can get help of Vader sentiment from this; https://towardsdatascience.com/sentimental-analysis-using-vader-a3415fef7664

```python
import nltk
import re
# nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import stop_words

sid = SentimentIntensityAnalyzer()
```

```python
c
```

```python
cleantext = []
for i in c:
    clean = re.sub("[^-9A-Za-z ]", "" , i)
    cleantext.append(clean)
cleantext
```

```python
df['Review text'] = cleantext
df
```

```python
df['scores'] = df['Review text'].apply(lambda review: sid.polarity_scores(review))
df
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
df
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
df
```

```python
from selenium.common import exceptions  

query = 'mobile covers'
query = query.replace(' ','%20')
url2 = f'https://www.daraz.pk/catalog/?from=input&q={query}'
```

```python
items = driver.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')
add = 1
tg = " "
tag = []
name = []
price = []
stars = []
reviews = []
country = []
count = 1

while add != 3:
    for i in items:
        
        try:
            t = i.find_element(By.CLASS_NAME, 'tag--huSJx')
            v = t.find_element(By.TAG_NAME, 'i')
            tg = "Daraz Mall"
        except StaleElementReferenceException as Exception: 
            tg = "Daraz Mall"
        except NoSuchElementException as Exception:
            tg = "Private"
            
        tag.append(tg)
            
        
        try:
            n = i.find_element(By.CLASS_NAME, 'title--wFj93').text
        except StaleElementReferenceException as Exception:
            n = i.find_element(By.CLASS_NAME, 'title--wFj93').text

        name.append(n)
        time.sleep(0.5)
        
        try:
            p = int(i.find_element(By.CLASS_NAME, 'price--NVB62').text[3:].replace(',',''))
        except StaleElementReferenceException as Exception:
            p = int(i.find_element(By.CLASS_NAME, 'price--NVB62').text[3:].replace(',',''))
        price.append(p)
        time.sleep(0.5)
        
        try:
            r = int(i.find_element(By.CLASS_NAME, 'rating__review--ygkUy').text[1:-1])
        except StaleElementReferenceException as Exception:
            r = int(i.find_element(By.CLASS_NAME, 'rating__review--ygkUy').text[1:-1])
        reviews.append(r)
        
        try:
            c = i.find_element(By.CLASS_NAME, 'rateAndLoc--XWchq').text[5:]
        except StaleElementReferenceException as Exception:
            c = i.find_element(By.CLASS_NAME, 'rateAndLoc--XWchq').text[5:]
        country.append(c)

            

        time.sleep(10)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    try:
        aa = driver.find_element(By.CLASS_NAME, 'pager--mb0Ws')
        bb = aa.find_elements(By.TAG_NAME, 'li')[2].click()
    except NoSuchElementException as Exception:
        aa = driver.find_element(By.CLASS_NAME, 'pager--mb0Ws')
        bb = aa.find_elements(By.TAG_NAME, 'li')[2].click()
    add += 1

df2 = pd.DataFrame({"Name": name,"Price": price, "Reviews": reviews, "Country": country, "Daraz Mall/Private": tag})
    
```
